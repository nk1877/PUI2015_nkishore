```python
alldata = sm.datasets.macrodata.load_pandas().data
```

```python
alldata.plot(figsize=(12,8));
```

```python
fig = pl.figure =(10,30)
ax= []
for i, data in enumerate(alldata.values.T):
    x2=0 if i%2 ==0 else 2
    ax.append(pl.subplot2grid(7,3),((i/2+x2))
    ax[i].plot(data,'k',alpha=0.5)
    ax[i].axis('off')
    try:
        ax[i].plot(np.where(data==max(data))[0],max(data),'o')
    except ValueError:
        ax[i].plot(np.where(data==max(data))[0][0],max(data),'o')
    try:
        ax[i].pl.plot(np.where(data==min(data))[0],min(data),'o')
    except ValueError:
        ax[i].plot(np.where(data==min(data))[0][0],min(data),'o')
    ax[i].text(0.8,0.5,"%.if"%(min(data)),fontsize=15,transform=ax[i].transAxis,color='IndianRed')
    ax[i].text(0.8,0.5,"%.if"%(max(data)),fontsize=15,transform=ax[i].transAxis,color='IndianRed')
```
