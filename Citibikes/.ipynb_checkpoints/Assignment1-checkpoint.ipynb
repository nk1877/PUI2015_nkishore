```python
import pylab as pl
import pandas as pd
import numpy as np
%pylab inline

import os

#this makes my plots pretty! but it is totally not mandatory to do it
#import json
#s = json.load( open(os.getenv ('PUI2015')+"/fbb_matplotlibrc.json") )
```

```python
df=pd.read_csv('/Users/nikhilkishore/Dropbox/nyuclasses/PUI/201504-citibike-tripdata.csv')
print df.columns
```

```python
df['ageM'] = 2015-df['birth year'][(df['usertype'] == 'Subscriber') & (df['gender'] == 1)]
df['ageF'] = 2015-df['birth year'][(df['usertype'] == 'Subscriber') & (df['gender'] == 2)]
```

```python
bins = np.arange(10, 99, 10)
df.ageM.groupby(pd.cut(df.ageM, bins)).agg([count_nonzero]).plot(kind='bar')
df.ageF.groupby(pd.cut(df.ageF, bins)).agg([count_nonzero]).plot(kind='bar')
```

```python
diM=df.ageM.groupby(pd.cut(df.ageM, bins)).agg([count_nonzero])
diF=df.ageF.groupby(pd.cut(df.ageF, bins)).agg([count_nonzero])

csM=df.ageM.groupby(pd.cut(df.ageM, bins)).agg([count_nonzero]).cumsum()

csF=df.ageF.groupby(pd.cut(df.ageF, bins)).agg([count_nonzero]).cumsum()

print np.abs(csM / csM.max()-csF / csF.max())

pl.plot(bins[:-1] + 5, csM / csM.max(), label = "M")
pl.plot(bins[:-1] + 5, csF / csF.max(), label = "F")
pl.legend()
```

```python
import scipy.stats
```

```python
scipy.stats.ks_2samp(df.ageM,df.ageF)
```

```python
###K stats is 0.50452949922592316 and p-value is 0.0 
###I will be using α as 0.05 and c(α) will be 1.36
```

```python
ps = scipy.stats.pearsonr(csM, csF)
```

```python
print "{} is correlation value and {}is pvalue" .format(ps[0][0],ps[1][0])
```

```python
sm=scipy.stats.spearmanr(diM, diF)
```

```python
print "{} is correlation value and {}is pvalue" .format(sm[0],sm[1])
```

```python
df['date'] = pd.to_datetime(df['starttime'])
```

```python
df['drider'] = 2015-df['birth year'][(df['usertype'] == 'Subscriber') & (df['date'].dt.hour>8) & (df['date'].dt.hour<23)]
df['nrider'] = 2015-df['birth year'][(df['usertype'] == 'Subscriber') & (df['date'].dt.hour >0) & (df['date'].dt.hour<8)]
```

```python
bins = np.arange(10, 99, 10)
df['drider'].groupby(pd.cut(df['drider'], bins)).agg([count_nonzero]).plot(kind='bar')
df['nrider'].groupby(pd.cut(df['nrider'], bins)).agg([count_nonzero]).plot(kind='bar')
```

```python
D=df.drider.groupby(pd.cut(df.drider, bins)).agg([count_nonzero])
N=df.nrider.groupby(pd.cut(df.nrider, bins)).agg([count_nonzero])

csD=df.drider.groupby(pd.cut(df.drider, bins)).agg([count_nonzero]).cumsum()

csN=df.nrider.groupby(pd.cut(df.nrider, bins)).agg([count_nonzero]).cumsum()

print np.abs(csD / csD.max()-csN / csN.max()).dropna()

pl.plot(bins[:-1] + 5, csD / csD.max(), label = "D")
pl.plot(bins[:-1] + 5, csN / csN.max(), label = "N")
pl.legend()
```

```python
import scipy.stats
```

```python
ks=scipy.stats.ks_2samp(df.drider, df.nrider)
```

```python
ps = scipy.stats.pearsonr(csD.dropna(), csN.dropna())
```

```python
print "{} is correlation value and {}is pvalue" .format(ps[0][0],ps[1][0])
```

```python
sm=scipy.stats.spearmanr(D, N)
```

```python
print "{} is correlation value and {}is pvalue" .format(sm[0],sm[1])
```
